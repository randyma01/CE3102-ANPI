<h2>
    <span>
        <strong>1.1 Matrices</strong> 
    </span>
</h2>
</span>

<span style="font-size: 12.0pt; line-height: 150%">

Todo sistema de ecuaciones, por ejemplo;

$$
\left.
2x_1 + x_2  = 1 \atop
x_1 + x_2 = 4
\right\}
$$

tiene una representación matricial de la forma:
    
$$
\begin{equation*}
\begin{pmatrix}
2 & 1 \\
1 & 1 \\
\end{pmatrix}
\times
\begin{pmatrix} 
x_1 \\
x_2 
\end{pmatrix}
=
\begin{pmatrix} 
1 \\
4 
\end{pmatrix}
\end{equation*}
$$

Donde: 
    
  - La primera matriz (llamada matriz de coeficientes) representa los coeficientes de la ecuación. 
    Esta se representa con la letra:
    **A**

    
    
  - La segunda matriz (llamada matriz de incógnitas) representa los valores a determinar. 
     Esta se representa con la letra:
    **x**
    
    
    
  - La tercera matriz (llamada matriz de términos independientes) representa la igualdad de cada ecuación.
     Esta se representa con la letra:
    **b**

De las matrices anteriores, se puede obtener la ecuación:

$$ 
\begin{equation*}
Ax = b
\end{equation*}
$$
    
Esto es, para toda matriz de tamaño MxM, es decir: misma cantidad de columnas y filas. Asimismo, la cantidad de filas para la matriz de términos independientes.  
    
En otras palabras, el número de ecuaciones independientes y el número de incógnitas deben ser las mismas. 

</span>

<a id="jac"></a>
<h2>
    <span>
        <strong>1.2 Solución de Matrices: Jacobi</strong> 
    </span>
</h2>
</span>

<span style="font-size: 12.0pt; line-height: 150%">

Hay múltiples formas de resolver la ecuación anterior. Un enfoque es utilizar iteraciones para aproximar la solución. Conmunmente, se utilizan sumas sucesivas. Los métodos iterativos tienen a seguir el siguiente esquema:
    
La matriz A se puede descomponener así:
     
$$
\begin{equation*}
A = L + D + U
\end{equation*}
$$
    
donde: 
    
  -  **L** es la matriz de A, pero únicamente con los valores debajo de la diagonal
    
    
  -  **D** es la matriz de A, pero únicamente con los valores de la diagonal
    
    
  -  **U** es la matriz de A, pero únicamente con los valores arriba de la diagonal
    
El método de Jacobi se basa en la siguiente expresión:
    
$$
\begin{equation*}
A = M - N
\end{equation*}
$$

donde, 
    
$$
\begin{equation*}
M = D  \\
N = -(L + U)
\end{equation*}
$$

    
Ahora, partiendo de la ecuación:
    
$$
\begin{equation*}
Ax = b 
\end{equation*}
$$
    
sustityendo **M** y **N** en **A** se obtiene la igualdad de:
    
$$
\begin{equation*}
(M + N)x = b \\
Mx + Nx = b 
\end{equation*}
$$

En este punto, se desea dejar una variable **x** en cada lado del igual. Por lo que se va a despejar el **M** de la siguiente manera:
    
$$
\begin{equation*}
Mx = b - Nx 
\end{equation*}
$$

Y ahora, se despeja **x**, obteniendo la ecuación final de: 
    
$$
\begin{equation*}
x = M^{-1}(b - Nx) 
\end{equation*}
$$
    
Al ser una iteración, la fórmula final, (*), seria:

    
$$
\left.
x^{(k+1)} = M^{-1}(b - Nx^{(k)}) \atop
x^{0} \in R^{n}
\right\}
$$   
    
    
Donde, el vector 

$$
x^{0} \in R^{n}
$$

corresponde a vector inicial, está compuesto de ceros y es de tamaño de la cantidad de incógnitas del sistema de ecuaciones.
 
Además, 

$$
x^{k}
$$

representa el valor anterior calculado, iniciando con el vector inicial y, 
    
$$
x^{k+1}_j
$$

representa el último, k-ésimo, valor calculado. 
    
    
La fórmula (*), al ser una iteración, se puede moldear como una serie, la cual se puede representar de la siguiente forma:
    
$$
\begin{equation*}
x^{(k+1)} = \dfrac{1}{a_{ii}} (b_i - \sum_{i=1}^{k} n_{ij}x_j^{k}) \\
i = 1, 2, 3, ...  
\end{equation*}
$$

Donde: 
    
- **M** es la matriz de diagonales de la matriz **A** y se representa como:
    
$$ 
\dfrac{1}{a_{ii}}
$$
    
    
- **b** es el vector de términos independientes y se representa como:
    
$$ 
b_{i}
$$
    
- **N** es la matriz sin diagonales de la **A** como:

$$ 
n_{ij}
$$
    
- **x** es el vector resultante de la suma anterior. Este inicial con el vector de valores de cero, y se representa como:
    
$$
x^{k}_{j}
$$

    
Esta es la forma no matricial de representación de la solución del sistema ecuaciones. Esta misma forma que se utiliza en la implementación computacional del método, que se muestran más adelante en la secciones [4. Código: C++ (v17)](#cp)

<h2>
    <span>
        <strong>2.1 Convergencia</strong> 
    </span>
</h2>
</span>

<span style="font-size: 12.0pt; line-height: 150%">

El método de Jacobi tiene una convergencia hacia las matrices diagonalmente dominantes.
    
Una matriz diagonalmente dominante es aquella donde el valor absoluto del elemento de la diagonal es mayor a la suma de lso valores absolutos de los demás elementos de una fila dada.
    
Si se aplica el método de Jacobi a una matriz con esta característica, siempre se va a obtner un cálculo aproximado de la solcuión al sistema.
    
En caso de que la matriz no sea diagonalmente dominante, puede que converga a la solución así como pueda que diverga.
    
Es por eso que al convergencia tiende a estas matrices.

</span>

<span style="color: #838383">
<a id="cri"></a>
<h2>
    <span>
        <strong>2.2 Criterio de Parada</strong> 
    </span>
</h2>
</span>

<span style="font-size: 12.0pt; line-height: 150%">

El métood de Jacobi al ser un método iteravo, va a repetirse una cantidad "n" de veces, hasta llegar a su criterio de parada, el cual es:
    
$$
error absoluto < tolerancia
$$

donde la **tolerancia**, que es dada, debe ser mayor al **error absoluto**, donde este es la norma de la expresión:
    
$$
Ax^{k} - b
$$

Por lo tanto, de manera formal, el criterio de parada es:
 
$$
||Ax^{(k)} - b|| < tol
$$
    
La norma a aplicar para el cálculo del valor absoluto queda a decisión subjetiva. Se recomienda utilizar la norma 2, también llamada norma Euclidiana. 

</span>

<a id="dvs"></a>
<h2>
    <span>
        <strong>2.3 Ventajas y desventajas</strong> 
    </span>
</h2>
</span>

<span style="font-size: 12.0pt; line-height: 150%">

**Ventajas**:
    
- Realiza pocas operaciones matriciales.
    
    
- El margen de error puede llegar a ser muy pequeño.
    
    
- De no requerir mucha presición, pueden hacer pocas iteraciones, reduciendo el proceso. 


**Desventajas**:
- Al ser un metodo iterativo puede consumir muchosos recursos, convirtiéndolo computacionalmente pesado.
    
    
- Su convergencia es estricta, debido a que en muchos casos diverge.
    
    
- El tiempo de cálculo y la precisión del resultado pueden depender de la elección de ciertos parámetros       como: la tolerancia.

```Aclaración: Para este trabajo, se basó en los códigos encontrados en las referencias [1] y [2]. Se tomó como guías esas implementaciones investigadas y se les hicieron los ajustos pertinentes para esta evaluación. Todos los derechos a sus respectivos autores.```

Este consiste en dos funciones:
    
- **Jacobi**: Realiza el cálculo de la solución de un sistema de ecuaciones
    
    
- **SumaJacobi**: Realiza el cálculo de la suma según la serie explicada en: [1. Análisis Matemático](#ac). 
    

- **Norma**: Realiza el cálculo de la norma 2. 
    
    
Para demostración su funcionamiento, se va a calcular la solución del siguiente sistema de ecuaciones:

$$
\begin{equation*}
\begin{pmatrix}
5 & 1 & 1 \\
1 & 5 & 1 \\
1 & 1 & 5 \\
\end{pmatrix}
\times
\begin{pmatrix} 
x_1 \\
x_2 \\
x_3 \\
\end{pmatrix}
=
\begin{pmatrix} 
7 \\
7 \\
7
\end{pmatrix}
\end{equation*}
$$

donde su solución es la matriz columna:

$$
\begin{equation*}
\begin{pmatrix}
x_1 \\
x_2 \\
x_3 \\
\end{pmatrix}
=
\begin{pmatrix} 
-1 \\
-1 \\
-1 \\
\end{pmatrix}
\end{equation*}
$$
    
</span>

````c++
#include <stdio.h>
#include <stdlib.h>

#define n 10

using namespace std;
````

````c++
/**
 * Funcion que realiza el calculo del vector resultante 
 * mediante la formula de la serie de jacobi.
 *
 *
 * @param vector_d vector de las diagonales
 * @param vector_prev vector resultante anterior
 * @param fila entero de la fila que se esta reccoriendo 
 * @param dim entero de la dimension de la matriz
 *
 * @return suma resultante de la serie
 */
float SumaJacobi(float vector_d[], float vector_prev[], int fila, int dim){
    
    // declaracion: valor resultante //
    float suma = 0;
    
    // declaracion: variable del contador //
    int i;

    // itereacion: realiza la suma de la serie de la formula de jacobi //
    for(i = 0; i < dim; i++){
        if (fila!= i){
            suma = suma + vector_d[i] * vector_prev[i];
        }
    }

    // retorno: valor resultante //
    return suma;
}
````

````c++
/**
 * Funcion que realiza el calculo de la norma de dos 
 * vectores dados utilizando la norma 2. %
 *
 *
 * @param vector_uno vector uno o A
 * @param vector_dos vector dos o B
 * @param dim entero de la dimension de la matriz
 *
 * @return norma de los vectores uno (A) y dos (B)
 */
float Norma(float vector_uno[], float vector_dos[], int dim){

    // declaracion: valor resultante //
    float norma = 0;

    // declaracion: variable del contador //
    int i;

    // itereacion: realiza el calcula de la norma por medio de una serie //
    for(i = 0; i < dim ; i++){
        // calculo: de la norma de los vectores resultante (anterior y actual)//
        norma = norma + (vector_uno[i] - vector_dos[i]) * (vector_uno[i] - vector_dos[i]);
    }

    // retorno: valor resultante //
    return norma;
}
````

````c++
/**
 * Esta funcion encuentra la solucion a un sistema 
 * de ecuaciones representado mediante la expresion:
 * A * x = b, utilizando el metodo de Jacobi.
 *
 *
 * @param A matriz de coeficientes
 * @param b vector de terminos independientes
 * @param dim dimension de la matriz
 *
 */
void Jacobi(float A[n][n], float b[n], int dim){

    // declaracion: contadores para los recorridos //
    int i, j;
    
    // declaracion: numero de iteraciones realizadas //
    int iteraciones = 0;

    // declaracion: tol (10^-8) //
    float tol = 0.0000000010;

    // declaracion: vector de error //
    float err = tol + 1;
  
    // declaracion: vector resultante //
    float x[dim];
    
    // declarcion: vector del calculo anterior //
    // este vector contiene los valores iniciales //
    float x_prev[dim];
    
    // declaracion: vector de los elementos de las diagonales //
    float aux[dim];

    // iteracion: mientras el error sea mayor que la //
    // tolerancia se ejecuta la serie que realiza el //
    // el calculo del vector resultante utilizando   //
    // la formula no matricial de Jacobi.            //
    while (tol < err){
        
        // iteracion: declaracion: del vector de la digonal de la matriz A //
        for(i = 0; i < dim; i++){
            for(j = 0; j < dim; j++){
                aux[j] = A[i][j];
            }
            
            // calculo: vector resultante mediante la formula de jacobi //
            x[i] = (1/A[i][i]) * (b[i] - SumaJacobi(aux, x_prev, i, dim));
        }
        
        // calculo: calcula de la norma del vector^ anterior y el vector^ actual //
        // ^ = vector resultante iniciando con los valores iniciales de 0's
        err = Norma(x, x_prev, dim);
    
        // iteracion: para mostrar en pantalla los resultados calculados //
        printf("\n\n Iteracion %d: \n", iteraciones);
        for(i = 0; i < dim;i ++){
            x_prev[i] = x[i];
            printf("X(%d)=%f \n", i, x[i]);
        }

        // aumento del contador de iteraciones realizadas //
        iteraciones++;
    }

    // iteracion: para mostrar en pantalla los resultados calculados //
    printf("Solucion del sistema\n");
    printf("Numero de iteraciones: %d \n", iteraciones);
    for(i = 0; i < dim; i++){
        printf("x(%d)=%f\n", i, x[i]);
    }
}
````

````c++
int dim = 3;

float A[n][n] = {
     {5, 1, 1},
     {1, 5, 1},
     {1, 1, 5}
};

float b[n] = {7, 7, 7};


Jacobi(A, b, dim);
````
