<span style="font-size: 12.0pt">

Implemente una script para encontrar un cero de la función
    
utilizando una tolerancia de 10^−10, mediante el método modificado de la bisección, con n = 10. Para esto, elabore la función 
    
```x = metodo_biseccion_mod(a,b,n,tol),``` 
    
donde *a*, *b* son los valores del intervalo, *tol* es una tolerancia dada para el criterio de parada, *n* es la cantidad de sub-intervalos en los que se dividirá el intervalo [a, b] y x es la aproximación a la solución de la ecuación. 
    
La función se detendrá cuando se cumpla la condición 
   
$$ 
|f1(x^{(i)}| < tol
$$
                     
donde x(i) es la i−ésima iteración del método modificado de la bisección.

- Nota 1: La función debe verificar que existe un cero en el intervalo seleccionado.
    
- Nota 2: La función f(x) tiene un cero en el intervalo [0,2].

</span>

````octave
%
%
% Funcion: Metodo de la Biseccion Modificado
%
%

function x = metodo_biseccion_mod(f1, a, b, n, tol)
  
  % error segun la tolerancia dada %
  error = tol + 1; 

  % division en n intervalos creando una %
  % iteracion de n repiticiones %   
  for( i = 0 : n)

    % verificacion de que haya solución en ese intervalo %
    if (f1(a) * f1(b) <= 0)

      % mientras el error sea mayor que la tolerancia dada %
      while (tol < error) 

        % metodo convencional de la biseccion %
        x = (a + b) / 2;
        if (f1(a) * f1(x) <= 0) 
          b = x;
        else 
          a = x;
        end  
        error = abs(f1(x));
      end
    else
      display('No existe solución en este intervalo.')
    end 
    
    % nuevos limites de intervalos a y b%
    b = i/b + i;
    a = b;
  end
end
````

````octave
%
%
% Script: Metodo Biseccion Modificado
%
%

% funcion %
f = @(x) sin(x)^2 - x^2 + 1

% intervalo %
a = 0;
b = 2;

% numero de iteraciones %
n = 10

% toleracia %
tol = 10^-10

x = metodo_biseccion_mod(f, a, b, n, tol)
````

Implemente en un una script el m ́etodo de Gauss-Seidel para resolver el sistema de ecuaciones 

$$
\begin{equation*}
Ax = d
\end{equation*}
$$
    
donde, 
    
$$
A ∈ R100×100, d ∈ R100
$$

- Nota: La implementación debe realizarse como se explic[o en clases. En el caso de no seguir el algoritmo explicado en clases, entonces el puntaje obtenido en esta pregunta será de 0.

</span>

````octave
%
%
% Funcion: Verifica que sea Diagonalmente Dominante
%
%

function bool = vf_diag_dom(A, m)

    % declaracion: valor resultante %
    bool = 0;

    % iteracion: recorrido de la matriz, mientras %
    % se verifica por filas que la matriz sea     %
    % diagonalmente dominante                     %
    for (i = 1 : m)
        % declaracion: elemento de la diagonal %
        ed = abs(A(i, i));
        
        % declaracion: suma de los elementos de una %
        % fila excluyendo el elemento diagonal      %
        suma = sum(abs(A(i, :))) - ed;
        
        % verificacion: siempre que el elemento diagonal %
        % sea mayor sigue, en caso contrario se sale del %
        % del ciclo                                      %
        if (suma < ed)
            bool = 1;
            continue;
        else
            bool = 0;
            break;
        end
    end
end
````

````octave
%
%
% Funcion: Realiza Sustitucion Hacia Adelante
%
%

function x = sust_adelante(A, b)
    
    % dimension del vector b %
    n = length(b); 
    
    % vector de zeros de la dimension de b %
    x = zeros(n, 1);
    
    % primer valor encontrado %
    x(1) = b(1) / A(1,1);

    % iteracion de la solucion hacia adelante %
    for ( i = 2 : n)
        suma = 0;
        for  (j = 1 : (i - 1))
            suma = suma + A(i, j) * x(j);
        end
        x(i) = (b(i) - suma) / A(i, i);
    end
end
````

````octave
%
%
% Funcion: Realiza Sustitucion Hacia Atras
%
%

function x = sust_atras(A, b)
    
    % dimension del vector b %
    n = length(b); 
    
    % vector de zeros de la dimension de b %
    x = zeros(n, 1);
    
    % primer valor encontrado %
    x(n) = b(n) / A(n,n);
    
    for (i = n : -1 : 1)
        suma = 0;
        for (j = i + 1 : n)
            suma = suma + A(i, j) * x(j);
        end
        x(i) = (b(i) - suma) / A(i, i);
    end
end
````

````octave
%
%
% Script: Metodo de la Gauss-Seidel
%
%

%% EJERCICIO INCOMPLETO: FALTO LA CREACION DE LA MATRIZ 100x100 CON EL VECTOR UNICAMENTE%%
%% EL METODO DE GAUSS-SEIDEL ESTA COMPLETO %%

% vector resultante %
xk = [];

% dimensiones de la matriz A %
[m, n] = size(A);

% verificiones que la matriz debe cumplir %
if (m != n)                    
    display("La matriz no es cuadrada.")
    return;
elseif (det(A) == 0)               
    display("La matriz no es invertible.")
    return;
elseif (!vf_diag_dom(A, m))
    display("La matriz no es diagonalmente dominante.")
    return;
elseif (n != size(b, 1))           
    display("El vector de valores independientes no coincide.")
    return;
else
    % vector inicial de ceros%
    x0 = zeros(m, 1);
    
    % asignacion vector resultante %
    xk = x0;

    % matriz M = (L + D)%
    M = tril(A);
    
    % matriz N = (-U) %
    N = A  - tril(A)

    % inversa de la matriz M%
    d = diag(M);
    Minv = diag(1./d);

    % solucion del sistema Mz = b%
    z = sust_adelante(Minv, b)

    % tolerancia dada %
    tol = 10^-8;

    % error segun la tolerancia dada %
    err = tol + 1;;

    while(tol < err)
        % calculo del nuevo b %
        b = N * xk;

        % solucion del sistema My = b
        yk = sust_atras(Minv, b);

        % formula de gauss-seidel %
        xk = yk + z; 

        % calculo del error absoluto mediante la norma 2 %
        err = norm(A * xk - b); 
    end
end
````
